```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
def linear_function(x):
    # 4x - 5y = 10
    # 5y = 4x - 10
    #  y = 4/5x - 2
    return 0.8 * x - 2

# y = x * 2
x_values = np.linspace(-15, 20, 20)
y_values = linear_function(x_values)
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values, label="x * 2", color="green")
plt.title("y = 0.8 * x - 2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
