```python
def linear_function1(x):
    # 4x - 5y = 10
    # 5y = 4x - 10
    #  y = 4/5x - 2
    return 0.8 * x - 2

def linear_function2(x):
    # 4x - 5y = 10
    # 5y = 4x - 10
    #  y = 4/5x - 2
    return -5 * x + 4

# y = x * 2
x_values = np.linspace(-15, 20, 20)
y_values1 = linear_function1(x_values)
y_values2 = linear_function2(x_values)
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green")

#plt.plot(x_values, y_values2, color="blue")

plt.title("y = 0.8 * x - 2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#plt.xticks([-15, -12, -5, 0, 5, 7, 12, 14])

# x_intersection = 6 / 5.8
# y_intersection = linear_function1(x_intersection)  
# plt.scatter(x_intersection, y_intersection, color="red", s=100, edgecolors="black", label="Intersection")
# plt.text(x_intersection + 1, y_intersection, f"({x_intersection:.2f}, {y_intersection:.2f})", fontsize=12, color="red")


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the equation
# 4x - 5y = 10
equation = sp.Eq(4*x - 5*y, 10)

# Solve for y dynamically
y_expr = sp.solve(equation, y)[0]  # Solve for y
y_expr

```

```python
def linear_function1(x):
    return 12 - x

def linear_function2(x):
    return x - 4

'''
x + y = 12
x - y = 4
'''
# y = x * 2
x_values = np.linspace(-100, 200, 50)
y_values1 = linear_function1(x_values)
y_values2 = linear_function2(x_values)
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green")
plt.plot(x_values, y_values2, color="blue")

plt.title("y = 0.8 * x - 2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#plt.xticks([-15, -12, -5, 0, 5, 7, 12, 14])

x_intersection = 8
y_intersection = linear_function1(x_intersection)  
plt.scatter(x_intersection, y_intersection, color="red", s=100, edgecolors="black", label="Intersection")
plt.text(x_intersection + 1, y_intersection, f"({x_intersection}, {y_intersection})", fontsize=12, color="red")


plt.show()
```
